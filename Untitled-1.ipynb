```python
url = "C:/Users/Jaskeerat Singh/Downloads/chromedriver_win32/chromedriver.exe"
driver = webdriver.Chrome(ChromeDriverManager().install())
driver.get('https://www.picturethisai.com/wiki')
```

```python
d = {}
```

```python
wait = WebDriverWait(driver,10)
plants = driver.find_elements(By.CLASS_NAME,"specie")
for i,plant in enumerate(plants):
    plantName = wait.until(EC.presence_of_element_located((By.CLASS_NAME,"specie")))
    print(plant.text)
    d[plant.text] = {}
    plant.click()
    careGuideText = wait.until(EC.presence_of_all_elements_located((By.CLASS_NAME,"basic-information-item-content-title")))
    careGuideTitle = wait.until(EC.presence_of_all_elements_located((By.CLASS_NAME,"basic-information-item-content-text")))
    print([x.text for x in careGuideText])
    print([x.text for x in careGuideTitle])
    # d[plant.text]['text'] = []
    # d[plant.text]["text"] += [x.text for x in careGuideText]
    # d[plant.text]["title"] = [x.text for x in careGuideTitle]
    specs = driver.find_elements(By.CLASS_NAME,"care-guide-items")
    if len(specs)!=0:
        # d[plant.text]["spec"] = specs[0].text
        print(specs[0].text)
    driver.back()
    print(i,"-----------------")
```
